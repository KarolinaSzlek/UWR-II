```python
#load df
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data')
df = df.rename({'p': 'target'}, axis=1)
```

```python
t_gini = Tree(df, criterion='gini')
t_gini.draw()
```

```python
t_entr = Tree(df, criterion='infogain')
t_entr.draw()
```
